```python
import pandas as pd
from s3fs.core import S3FileSystem
from functools import reduce

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
s3 = S3FileSystem(anon=False)   #type = s3fs.core.S3FileSystem
key = '2016_2017_summary.xlsx'
bucket = 'comparek12-data'
```

```python
summary_2016_2017_df = pd.read_excel(s3.open('{}/{}'.format(bucket, key), mode='rb'))
```

```python
key = '2016_2017_refp.xlsx'
refp_df = pd.read_excel(s3.open('{}/{}'.format(bucket, key), mode='rb'))
refp_df.head()
```

```python
master_df = [summary_2016_2017_df, caasp_ela_df, caasp_math_df, college_readiness_df, demographics_df, other_performance_df, refp_df] 
#on=['District Name','CDS Number (District)'])
```

```python
df_final = reduce(lambda left,right: pd.merge(left,right,on='District Name'), master_df)
```
